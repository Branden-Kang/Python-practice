```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'userId': [1,1,2,2],
     'date': ['2020-01-01','2020-01-02','2020-01-01','2020-01-02'], 
     'value': [1,1,2, None]}
)

df.groupby('userId').apply(lambda x: x.ffill())
```

```python
df.groupby('userId', group_keys=False).apply(lambda x: x.ffill())
```

```python
df.groupby('userId', group_keys=True).apply(lambda x: x.ffill())
```

```python
df = pd.DataFrame(
    {'userId': ['Alice','Alice','Bob','Bob'],
     'date': ['2020-01-01','2020-01-02','2020-01-01','2020-01-02'], 
     'value': [1,1,2, None]}
    , index=['row-1', 'row-2','row-3','row-4']
)
```

```python
df.groupby('userId', group_keys=True).apply(lambda x: x.set_index('date'))
```
