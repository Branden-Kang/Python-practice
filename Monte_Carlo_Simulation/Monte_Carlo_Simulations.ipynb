```python
import random

def getDiceRollResult():
  die_1 = random.randint(1,6)
  die_2 = random.randint(1,6)

  if die_1 == die_2:
    result = True
  else:
    result = False

  return result
```

```python
SIMULATIONS = 10000
ROLLS_PER_GAME = 1000
BET_AMOUNT = 10
STARTING_BALANCE = 10000
WIN_PROBABILITY = list()
END_BALANCE = list()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
plt.title(f"Double Dice Game: {SIMULATIONS} Simulations")
plt.xlabel("Roll Number")
plt.ylabel("Balance Amount")
plt.xlim([0, ROLLS_PER_GAME])

for sim in range(SIMULATIONS):
  balance = [STARTING_BALANCE]
  roll_count = [0]
  win_count = 0

  while roll_count[-1] <= ROLLS_PER_GAME:
    game_result = getDiceRollResult()

    if game_result:
      balance.append(balance[-1] + (6 * BET_AMOUNT))
      win_count += 1
    else:
      balance.append(balance[-1] - BET_AMOUNT)
    
    roll_count.append(roll_count[-1] + 1)
  WIN_PROBABILITY.append(win_count/roll_count[-1])
  END_BALANCE.append(balance[-1])
  plt.plot(roll_count, balance)
#   plt.show()
```

```python
overall_win_probability = sum(WIN_PROBABILITY)/len(WIN_PROBABILITY)
overall_end_balance = sum(END_BALANCE)/len(END_BALANCE)
print(f"Average win probability after {SIMULATIONS} runs: {overall_win_probability}")
print(f"Average ending balance after {SIMULATIONS} runs: {overall_end_balance}")
```
