```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, HBox, Layout,Button,GridspecLayout
from IPython.core.display import display

plt.style.use('ggplot')

def sinwave(A, f, phi, color):
    # set up plot
    fig, ax = plt.subplots(figsize=(10,4))
    ax.set_ylim([-4,4])
    ax.grid(True)
    ax.set_xlabel('X')
    ax.set_ylabel('Sine Wave')
    ax.set_title('Sine Wave Widget for Interactive Plot')
    fig.canvas.toolbar_position='bottom' # set zoom-able and resize-able under the figure 
    fig.canvas.header_visible = False # Hide the Figure name at the top of the figure

    # generate x values
    x = np.linspace(0, 2*np.pi, 100)
    y = A*np.sin(x*f + phi);
    ax.plot(x,y,color)
```

```python
colors=['blue','red','orange']
widget = interactive(sinwave, A=(0,4,.1), f=(0,10,1), phi=(0,2,0.5),color=colors)
```

```python
widget.children
```

```python
controls=HBox(widget.children[:-1],layout=Layout(flex_flow='column wrap'), width='auto', height='auto', output=widget.children[-1])
output=widget.children[-1]
```

```python
def create_expanded_buttton(description, button_style):
    return Button(description=description,button_style=button_style,layout=Layout(height='auto',width='auto'))
```

```python
# use GIrdspecLayout to define the widget layout
grid = GridspecLayout(10, 6, height='500px')
grid[9,:] = create_expanded_buttton('Sine Web Plot App','success')
grid[2:8, 0:2] = controls
grid[1:9, 2:] = output
grid[9,:] = create_expanded_buttton('\xa9 2022 Copyright Branden','info')
grid
```
