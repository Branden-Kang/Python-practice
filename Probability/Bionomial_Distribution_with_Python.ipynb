# Probability Mass Function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success

# Values for which we want to calculate PMF
x = np.arange(0, n + 1)
# Calculate PMF
pmf_values = binom.pmf(x, n, p)

# Plot PMF
plt.bar(x, pmf_values, color='blue', alpha=0.7)
plt.xlabel('Number of successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF')
plt.show()
```

# Cumulative Distribution Function (CDF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom


# Parameters
n = 10  # Number of trials
p = 0.5  # Probability of success

# Values for which we want to calculate CDF
x = np.arange(0, n + 1)
# Calculate CDF
cdf_values = binom.cdf(x, n, p)

# Plot CDF
plt.plot(x, cdf_values, marker='o', linestyle='-', color='blue')
plt.xlabel('Number of successes')
plt.ylabel('Cumulative Probability')
plt.title('Binomial Distribution CDF')
plt.grid()
plt.show()
```

# Random Sampling

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10  # Number of trials (e.g., patients per experiment)
p = 0.7  # Probability of success (e.g., response rate to the drug)
num_samples = 1000  # Number of random samples to generate

# Generate random samples
samples = np.random.binomial(n, p, num_samples)

# Plot the results
plt.hist(samples, bins=range(n + 2), align='left', color='blue', alpha=0.7, edgecolor='black')
plt.xlabel('Number of successes')
plt.ylabel('Frequency')
plt.title('Random Sampling from Binomial Distribution')
plt.grid(axis='y')
plt.show()
```
