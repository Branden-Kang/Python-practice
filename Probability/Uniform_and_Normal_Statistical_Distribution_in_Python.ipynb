# Example of Uniform Distribution in Python


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the uniform distribution
a = 0
b = 1

# Generate random numbers following a uniform distribution
uniform_data = np.random.uniform(a, b, 1000)

# Plot the histogram of the uniform data
plt.hist(uniform_data, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')

# Plot the theoretical PDF (perfect rectangular distribution)
x = np.linspace(a, b, 100)
y = [1/(b-a)] * len(x)
plt.plot(x, y, 'r--', linewidth=2)

# Add titles and labels
plt.title('Uniform Distribution between 0 and 1')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

# Example of Normal Distribution in Python


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate data
data = np.random.normal(mu, sigma, 1000)

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the theoretical PDF
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'r-', linewidth=2)

# Add titles and labels
plt.title('Normal Distribution (Mean=0, Std Dev=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
