```python
import pandas as pd
import numpy as np

train = pd.read_csv("http://bit.ly/kaggletest")
```

```python
train.head()
```

```python
train['Sex_num'] = train.Sex.map({'male':1,'female':0})
```

```python
train.loc[0:4,{'Sex','Sex_num'}]
```

# Apply function


```python
train['Name_length'] = train.Name.apply(len)
train.loc[0:4,{'Name','Name_length'}]
```

```python
train['Fare_ceil'] = train.Fare.apply(np.ceil)
train.loc[0:4,{'Fare','Fare_ceil'}]
```

```python
train.head()

def getelement(mylist,pos):
    return mylist[pos]
```

```python
train['LastName']  = train.Name.str.split(',').apply(getelement,pos =0)
```

```python
train.loc[0:4,{'Name','LastName'}]
```

```python
train.Name.str.split(',').apply(lambda x:x[0]).head()
```
