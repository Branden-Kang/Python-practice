```python
import pseudopeople
import pandas as pd

my_config = {
    'decennial_census': {
        'column_noise': {
            'first_name': {
                'leave_blank': {
                    'cell_probability': 0.06
                },
                'use_nickname': {
                    'cell_probability': 0.06
                },
                'use_fake_name': {
                    'cell_probability': 0.06
                },
                'make_phonetic_errors': {
                    'cell_probability': 0.06
                },
}
        }}}

df_1 = pseudopeople.generate_decennial_census(config = my_config, seed = 1)
df_2 = pseudopeople.generate_decennial_census(config = my_config, seed = 2)

x = [2000, 6000]
for size in x:
    sample_size = round(size * 1.5)
    sample_set = df_1.sample(sample_size)
    cut = round(sample_size / 3)

    dfA_first = df_1[0:cut]
    dfB_first = df_2[0:cut]

    dfA_last = sample_set[(cut):(2 * cut)]
    dfB_last = sample_set.tail(cut)

    frame_a = [dfA_first, dfA_last]
    frame_b = [dfB_first, dfB_last]
    dfA = pd.concat(frame_a)
    dfB = pd.concat(frame_b)

    dfA.to_csv(f"{size}_dfA.csv", index=False)
    dfB.to_csv(f"{size}_dfB.csv", index=False)
```
