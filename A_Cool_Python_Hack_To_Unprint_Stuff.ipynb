```python
CURSOR_UP = '\033[1A'
CLEAR = '\x1b[2K'
CLEAR_LINE = CURSOR_UP + CLEAR

print('apple')
print('orange')
print('pear')

from time import sleep
sleep(1)

# clears TWO lines
print(CLEAR_LINE * 2, end='')
print('pineapple')
```

```python
CURSOR_UP = '\033[1A'
CLEAR = '\x1b[2K'
CLEAR_LINE = CURSOR_UP + CLEAR

from time import sleep
print()

for i in range(1, 101):
    sleep(0.02)
    print(CLEAR_LINE * 2, end='')
    print('[' + '#'*i + '-'*(100-i) + ']')
```
