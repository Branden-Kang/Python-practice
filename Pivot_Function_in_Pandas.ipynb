```python
import pandas as pd
df = pd.DataFrame({'A': ['one', 'one', 'two', 'two'], 

                   'B': ['a', 'b', 'a', 'b'], 

                   'C': [1, 2, 3, 4]}) 
df
```

```python
df.pivot(index='A', columns='B', values='C')
```
