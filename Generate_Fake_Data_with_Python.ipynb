```python
from faker import Faker
# Faker generator
fake_data  = Faker()
# make every generation create same output, optional
Faker.seed(6)
# generate English language name
print(fake_data.name())
# generate email address
print(fake_data.safe_email())
# generate home address
print(fake_data.address())
# generate personal profile
profile = fake_data.simple_profile()
for m,v in profile.items():
    print('{}: {}'.format(m, v))
# generate Chinese language name
fake_data2  = Faker('zh_CN')
Faker.seed(5)
print(fake_data2.name())
# generate Russian language country name
fake_data3  = Faker('ru_RU')
Faker.seed(5)
print(fake_data3.country())
# create CSV
import pandas as pd
names = []
for _ in range(0, 10):
    names.append(fake_data.name())
emails = []
for _ in range(0, 10):
    emails.append(fake_data.safe_email())
phones = []
for _ in range(0, 10):
    phones.append(fake_data.phone_number())
fake_df = pd.DataFrame({'Name': names,'Email': emails, 'Phone': phones})
print(fake_df)
data_csv = [fake_data.profile() for i in range(0, 10)]
df = pd.DataFrame(data_csv)
```
