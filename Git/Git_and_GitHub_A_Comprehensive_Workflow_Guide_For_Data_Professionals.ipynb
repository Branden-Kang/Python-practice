# Three brancheds
- main — production. Code here is live, stable, and deployable. You never commit directly to main.
- dev — integration. Features land here first for testing before they go to main. Think of it as the staging environment for your branch strategy.
- feature/feature_name — your daily workspace. One branch per feature, created from dev, merged back into dev when done.

```
git init                    # initialise a new git repo in your current folder
git checkout -b main        # create the main branch
git checkout -b dev         # create dev branch from main
git push -u origin dev      # push dev to GitHub and set it as the tracked upstream branch
```

# Building Feature by Feature
## Step 1: Create your feature branch from dev
```
git checkout dev                 # switch to dev — always branch off dev, not main
git checkout -b feature/extract  # -b creates a new branch and switches to it
```

## Step 2: Do the work, then stage and commit
```
git add src/extract.py                                          # stage this file for commit
git commit -m "feat: add raw data extraction from source API"  # commit with a clear message
```

## Step 3: Push your branch to GitHub
```
git push -u origin feature/extract  # -u sets the upstream; after this, git push is enough
```


## Step 4: Pull while you’re still building
```
git checkout feature/extract
git pull origin dev    # bring the latest dev changes into your feature branch
```


## Step 5: Open a Pull Request on GitHub — don’t merge locally
1. Go to your repository on GitHub
3. Click Compare & pull request (GitHub shows this automatically after a push)
3. Set base branch to dev and compare branch to feature/extract
4. Write a short description of what the feature does
5. Click Create pull request


# After the PR is approved and merged on GitHub, clean up the branch locally:


```
git checkout dev
git pull origin dev            # sync your local dev with the merged result on GitHub
git branch -d feature/extract  # delete the local branch — it's done its job
```

```
git push origin --delete feature/extract  # delete the remote branch from the command line
```

# Repeating the Cycle: feature/load and feature/transform
```
git checkout dev
git pull origin dev             # get the merged extract feature before branching
git checkout -b feature/load    # create the load feature branch from the updated dev

git add src/load.py
git commit -m "feat: load raw data into the data warehouse"
git push -u origin feature/load
# → open PR on GitHub: feature/load → dev
# → review and merge on GitHub
git checkout dev
git pull origin dev              # sync local dev after the PR is merged
git branch -d feature/load       # clean up the local branch

git checkout dev
git pull origin dev                   # pick up the load work before branching
git checkout -b feature/transform

git add src/transform.py
git commit -m "feat: add dbt transformations for staging layer"
git push -u origin feature/transform
# → open PR on GitHub: feature/transform → dev
# → review and merge on GitHub
git checkout dev
git pull origin dev
git branch -d feature/transform
```

# When dev Is Ready to Merge to main
```
git checkout main
git pull origin main   # your local main is now in sync with the merged result

git tag -a v1.0.0 -m "Release: complete ELT pipeline"
git push origin v1.0.0
```

# What About Bugs and Small Fixes?
```
git checkout dev
git pull origin dev                   # get the latest dev before branching
git checkout -b fix/extract-endpoint

# Make the fix...
git add src/extract.py
git commit -m "fix: add extra columns for extraction"
git push -u origin fix/extract-endpoint   # push to GitHub

# → open PR on GitHub: fix/extract-endpoint → dev
# → review and merge on GitHub

git checkout dev
git pull origin dev                            # sync local dev after merge
git branch -d fix/extract-endpoint             # remove local branch
git push origin --delete fix/extract-endpoint  # remove remote branch
```
