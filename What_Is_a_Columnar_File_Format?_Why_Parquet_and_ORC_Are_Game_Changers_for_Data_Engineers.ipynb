```python
import pandas as pd

df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "country": ["US", "UK", "India"],
    "age": [29, 31, 27]
})

# Save to Parquet
df.to_parquet("users.parquet", engine="pyarrow", index=False)

# Read from Parquet
df2 = pd.read_parquet("users.parquet")
print(df2.head())
```
