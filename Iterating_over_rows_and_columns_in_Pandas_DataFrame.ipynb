```python
import pandas as pd
```

```python
dict_ = {'name': ["Ameena",'Muskan',"Needha","Sanvi"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score': [90, 40, 80, 98]}
df = pd.DataFrame(dict_)
df
```

```python
# iterating over rows using iterrows() function
for i, j in df.iterrows():
    print(i, j)
    print()
```

```python
for key, value in df.items():
    print(key, value)
    print()
```

```python
for i in df.itertuples():
    print(i)
```

```python
columns = list(df)

for i in columns:
    print(df[i])
```
