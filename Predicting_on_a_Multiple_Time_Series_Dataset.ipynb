```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/TracyRenee61/Multiple-Time-Series-Predictions/main/Sales_Transactions_Dataset_Weekly.csv')
data = data.filter(regex=r'Product|W')
data
```

```python
data.isnull().sum().sum()
```

```python
code = data.Product_Code

data = data.drop(['Product_Code'], axis=1)

#split the dataset in two
# the last column is our label
y_train = data.iloc[:,-1:].values
#drop last column of data
X_train = data.iloc[:, :-1]
#drop first colum of data
X_test = data.iloc[:,1:]
# lets have a look on the shape 
X_train.shape,y_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)
# Scale the train set
X_train = scaler.transform(X_train)
# Scale the validation set
#X_val = scaler.transform(X_val)
# Scale the test set
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPRegressor

model = MLPRegressor(max_iter=3000, solver='lbfgs', random_state=1, verbose=True).fit(X_train, y_train)
y_pred = model.predict(X_train)
y_pred = y_pred.astype(int)
y_pred
```

```python
print(model.score(X_train, y_train))
```

```python
prediction = model.predict(X_test)
prediction = prediction.astype(int)
prediction
```

```python
print ('Compare y_train to y_pred: {0}'.format((y_pred == y_train).sum().astype(float) / len(prediction)))
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_train, y_pred, squared=False)
rmse
```

```python
#create dataframe
output = pd.DataFrame({'code': code, 'sales': prediction})
#create csv
output.to_csv('submission.csv', index=False)
#upload submission
submission = pd.read_csv("submission.csv")
submission
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=3000, max_depth=15, random_state=1, verbose=True).fit(X_train, y_train)
r_pred = model.predict(X_train)
r_pred = r_pred.astype(int)
r_pred
```

```python
print(model.score(X_train, y_train))
```

```python
print ('Compare r_pred to y_train: {0}'.format((r_pred == y_train).sum().astype(float) / len(prediction)))
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_train, r_pred, squared=False)
rmse
```
