```python
def square_numbers():
  num = 1
  while num**2 <= 100:
    yield num**2
    num += 1
```

```python
def test_generator():
  num = 0
  while num < 100:
    yield num
    num += 1
my_list = [i for i in range(100)]
my_gen = test_generator()
```

# Create Infinite Sequences


```python
def natural_numbers():
  i = 0
  while True:
    yield i
    i += 1
```

```python
for num in natural_numbers():
  print(num)
```

```python
square_numbers = (i*i for i in range(10))
square_numbers
```
