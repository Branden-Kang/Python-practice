# 2. attrs – Lightweight, Elegant Data Classes

```python
import attr
@attr.s
class Config:
    host = attr.ib(type=str)
    port = attr.ib(type=int, default=5432)
```

# 9. beartype – Runtime Type Checking (Done Right)

```python
from beartype import beartype
@beartype
def add(a: int, b: int) -> int:
    return a + b
```
