```python
from itertools import product

a = [1,2]
b = [3,4]

prod = product(a,b)

print(list(prod))
```

```python
from itertools import permutations

a = [1,2,3]

perms = permutations(a)

print(list(perms))
```

```python
from itertools import combinations

a = [1,2,3]

combs = combinations(a, 2)

print(list(combs))
```

```python
from itertools import combinations_with_replacement

a = [1,2,3]

combs = combinations_with_replacement(a, 2)

print(list(combs))
```

```python
from itertools import accumulate

a = [1,2,3,4]

acc = accumulate(a)

print(list(acc))
```

```python
from itertools import accumulate
import operator

a = [1,2,3,4]

prods = accumulate(a, func=operator.mul)

print(list(prods))
```

```python
from itertools import groupby

a = [1,2,3,4,5]

group = groupby(a, key=lambda x: x < 3)

for k, v in group:
    print(k, list(v))
```

```python
from itertools import count

for i in count(10):
    print(i)
    if(i == 100):
        break
```

```python
from itertools import cycle

a = [1,2,3]
for i in cycle(a):
    print(i)
```

```python
from itertools import repeat

for i in repeat(10):
    print(i)
```
