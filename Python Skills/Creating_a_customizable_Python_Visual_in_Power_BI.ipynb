```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", palette="pastel")

if pokedex_df.iloc[0,0] == 'Boxplot':
    sns.boxplot(x="type", y="Attack", palette=["m", "g"],
                data=pokedex_df)
    sns.despine(offset=10, trim=True)
elif pokedex_df.iloc[0,0] == 'Scatterplot':
    sns.scatterplot(x="type", y="Attack",
                sizes=(1, 8), linewidth=0,
                data=pokedex_df)
elif pokedex_df.iloc[0,0] == 'Violinplot':
    sns.violinplot(data=pokedex_df, x="type", y="Attack", palette="Set3", bw=.2, cut=1, linewidth=1)
plt.show()
```
