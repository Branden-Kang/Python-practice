# 9. match with guards

````python
match amount:
    case x if x < 0:
        return "invalid"
    case 0:
        return "zero"
    case x if x > 10_000:
        return "large"
    case _:
        return "normal"
````

# 10. contextlib.ExitStack

````python
from contextlib import ExitStack

with ExitStack() as stack:
    files = [stack.enter_context(open(p)) for p in paths if p.exists()]
    return [f.read() for f in files]
````
