```python
raw = pd.read_csv('https://raw.githubusercontent.com/thihanaung-thnn/data_insight/main/notebooks/datasets/tb_raw.csv')
raw = raw.drop(['mu','fu'],axis=1)
raw
```

```python
df = raw.melt(id_vars=['country','year'],value_vars=raw.columns[3:], var_name='column', value_name='cases')  
df
```

```python
df['sex'] = df.column.str[0].str.upper()
df['age'] = df.column.str[1:]
df.age.unique()
```

```python
age = []
for n in range(len(df.age)):
    if len(df.age[n])==4:
        age.append(df.age[n][:2] + '-' + df.age[n][2:])
    elif len(df.age[n])==3:
        age.append(df.age[n][0] + '-' + df.age[n][1:])
    else:
        age.append(df.age[n] + '+')
df['age'] = age

df.dropna(inplace=True)
df.cases = df.cases.astype('int')
clean = df[['country','year','sex','age','cases']]
clean
```

```python
raw = pd.read_csv('https://raw.githubusercontent.com/thihanaung-thnn/data_insight/main/notebooks/datasets/weather_raw.csv')
raw
```

```python
df = raw.iloc[:,1:].melt(id_vars=['id','year','month','element'],var_name='day',value_name='temp')
df.dropna(inplace=True)
df['day'] = df.day.str[1:]
df['date'] = df.year.astype('str') + '/' + df.month.astype('str').str.zfill(2) + '/' + df.day.astype('str').str.zfill(2)
df['date'] = pd.to_datetime(df['date'])
clean = df.pivot(index= ['date'], columns='element',values='temp')
clean
```
