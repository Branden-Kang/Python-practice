# 3. Checking Free RAM

```python
import psutil
 
def bytes_format(n):
    for x in ['bytes', 'KB', 'MB', 'GB']:
        if n < 1024.0:
            return "%3.1f %s" % (n, x)
        n /= 1024.0

response=list(psutil.virtual_memory())

# Total ram.
print("Total Ram:", bytes_format(int(response[0])))

# Ram used.
print("Ram used :", bytes_format(int(response[3])))
 
# Ram free.
print("Ram Free :", bytes_format(int(response[1])))
```

# 4. Encrypt / Decrypt String

```python
!pip install cryptography
```

```python
# Encryption and Decryption of String

from cryptography.fernet import Fernet
 
key = Fernet.generate_key() 
cipher_suite = Fernet(key)
 
# Encoded Code
encoded_text = cipher_suite.encrypt(b"Welcome back Python Developer")
print("encoded text: ", encoded_text)
 
# Decoded Code
decoded_text = cipher_suite.decrypt(encoded_text)
print("decoded text: ", decoded_text)#Output:
# encoded text: b'gAAAAABhdbmXBR1_EuQ_HzcomFT.......BUX1HZjAqaE='
# decoded text: b'Welcome back Python Developer'
```

# 10. Resize an Image

```python
import cv2
 
img = cv2.imread('Image.jpg', cv2.IMREAD_UNCHANGED)

scale_percent = 50 # change accordingly
 
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
 
dimension = (width, height)

resized = cv2.resize(img, dimension, interpolation = cv2.INTER_AREA)

cv2.imwrite('resized.jpg', resized) # Image saved

cv2.waitKey(0)

cv2.destroyAllWindows()
```
