```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Generate data from a quadratic function with random noise e
def make_data(n):
  x = np.random.uniform(size=(n,1))
  e = np.random.normal(0, 0.1, size=x.shape)
  y = np.square(x) + x + 1 + e
  return (x,y)

x,y = make_data(100)
fig, ax = plt.subplots(dpi=100)
plt.scatter(x,y.ravel())
```

# Sine Function

```
# Generate data for sine function with random noise e
def generate_data(n):
  x = np.random.uniform(size=(n,1))
  e = np.random.normal(0,0.08,x.shape)
  y = np.sin(2*np.pi*x)+e
  return (x,y.ravel())

x,y = generate_data(100)
fig, ax = plt.subplots(dpi=100)
plt.scatter(x,y)
```

# Arctan Function

```
# Generate data for arctan function with random noise e
def generate_data(n):
  x = np.random.uniform(size=(n,1))
  e = np.random.normal(0,0.08,x.shape)
  y = np.arctan(2*np.pi*x)+e
  return (x,y.ravel())

x,y = generate_data(100)
fig, ax = plt.subplots(dpi=100)
plt.scatter(x,y)
```

# Exponential Function

```
# Generate data for natural exponential function with random noise e
def generate_data(n):
  x = np.random.uniform(0, 10, size=(n,1))
  e = np.random.normal(0,1000,x.shape)
  y = np.exp(x) + e
  return (x,y.ravel())

x,y = generate_data(100)
fig, ax = plt.subplots(dpi=100)
plt.scatter(x,y)
```

# Logarithmic Function

```
# Generate data for natural logarithm function with random noise e
def generate_data(n):
  x = np.random.uniform(0, 10, size=(n,1))
  e = np.random.normal(0,0.40,x.shape)
  y = np.log(x) + e
  return (x,y.ravel())

x,y = generate_data(100)
fig, ax = plt.subplots(dpi=100)
plt.scatter(x,y)
```
