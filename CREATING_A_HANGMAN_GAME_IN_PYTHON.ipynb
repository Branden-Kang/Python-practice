```python
#Import random module
import random
```

```python
#Create the graphics of the game
hang = ["""
H A N G M A N - Snacks Edition

   +---+
   |   |
       |
       |
       |
       |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
      |
      |
      |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
  |   |
      |
      |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
 /|   |
      |
      |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========""", """
H A N G M A N - Snacks Edition

  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
========="""]
```

```python
#Step1: Define a function to get a random word from a predefined list
def randomword():
  words = ['cookies', 'gummyworms', 'gummybears', 'chipsahoy', 'icecream', 'chocolate', 'pizza', 'fudge']

  word = random.choice(words)
  return word
```

```python
#Step2: lets 
def board(hang, lettersmissed, correctletters, secretword):
  #print the number of letters missed
  print(hang[len(lettersmissed)])
  print()

  print('Missed Letters:', end = '')
  for letter in lettersmissed:
    #for every letter in lettermissed, print out that letter
    print(letter, end = '')
  print('\n')
  #create blanks for every letter in that secretword
  blanks = '_' * len(secretword)
  #for every character in the secretword, replace the blanks with the correctly guessed letter
  for i in range(len(secretword)):
    if secretword[i] in correctletters:
      blanks = blanks[:i]  + secretword[i] + blanks[i + 1:]
  
  #show the secret word with spaces in between letters
  for letter in blanks:
    print(letter, end = '')

  print('\n')
```

```python
#Step3: Create a function that checks through the user's input
def guess(guessedword):
  while True:
    guesses = input("Guess a letter: ")
    guesses = guesses.lower()
    #if the length of your guess is not 1, prompt the player to enter the letter
    if len(guesses) != 1:
      print('Please, enter a single letter')
    elif guesses in guessedword:
      print("You have already guessed that letter, choose a different letter")
    elif guesses not in 'abcdefghijklmnopqrstuvwxyz':
      print("Please, enter a letter")
    else:
      return guesses
```

```python
#Step4: Ask the user if they want to play again:
def playagain():
  again = input('\nDo you want to play again?').lower().startswith('y')
  return again
```

```python
#Time for the game logic
#create empty strings for our missed letters and correct letters
missedletters = ''
correctletters = ''
secretword = randomword() #choose a random word 
gameisdone = False

while True:
  #while the game is true, display the hang, missedletters, correctletters, and the randomly picked word
  board(hang, missedletters, correctletters, secretword)

  guesses = guess(missedletters + correctletters)

  #if the guesses made by the player are in the secretword, concatenate that word to the correctletter string
  if guesses in secretword:
    correctletters = correctletters + guesses

    foundallletters = True
    #for every letter in the secretword:
    for i in range(len(secretword)):
      #if a letter is not in the correctletters, then you haven't found all the letters
      if secretword[i] not in correctletters:
        foundallletters = False
        break
    #if you found all letters, print the statement:
    if foundallletters:
      print("Yes, the secret word is '" + secretword + '" !You won!')
      gameisdone = True

  else:
    #if not, they are missed letters, so add them to the missed letters string
    missedletters = missedletters + guesses
    #if the number of missed letters is equal to the number of indexes in the hang list, the game is over
    if len(missedletters) == len(hang) - 1:
      board(hang, missedletters, correctletters, secretword)
      print("You ran out of guesses" + str(len(missedletters)) + 'missed guesses and' + str(len(correctletters)) + ' correct guesses, the word was' + secretword + '"')
      gameisdone = True
      
  #Reset the game
  if gameisdone:
    if playagain():
      missedletters = ''
      correctletters = ''
      gameisdone = False
      secretword = randomword()
    else:
      break
```
