# 1. Normal Distribution Confidence Interval

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Example data: exam scores
exam_scores = [75, 80, 85, 90, 92, 78, 88, 95, 79, 82]

# Calculating sample mean
sample_mean = np.mean(exam_scores)

# Calculating standard error
standard_error = np.std(exam_scores) / np.sqrt(len(exam_scores))

# Setting confidence level
confidence_level = 0.95

# Calculating Z-score
z_score = norm.ppf((1 + confidence_level) / 2)

# Calculating confidence interval
confidence_interval = (
    sample_mean - z_score * standard_error,
    sample_mean + z_score * standard_error
)

print(f"Sample Mean: {sample_mean}")
print(f"Standard Error: {standard_error}")
print(f"Z-Score: {z_score}")
print(f"Confidence Interval: {confidence_interval}")

# Plotting the data
plt.hist(exam_scores, bins=10, color='skyblue', alpha=0.7, label='Exam Scores')

# Plotting the confidence interval
plt.axvline(confidence_interval[0], color='red', linestyle='dashed', linewidth=2, label='Confidence Interval')
plt.axvline(confidence_interval[1], color='red', linestyle='dashed', linewidth=2)

plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Normal Distribution Confidence Interval')
plt.legend()
plt.show()
```

# 2. T-Distribution Confidence Interval

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# Example data: exam scores for a smaller class
small_class_scores = [75, 80, 85, 90, 92]

# Calculating sample mean
sample_mean_small = np.mean(small_class_scores)

# Calculating standard error using the sample standard deviation
standard_error_small = np.std(small_class_scores, ddof=1) / np.sqrt(len(small_class_scores))

# Setting confidence level
confidence_level_small = 0.95

# Calculating T-score
t_score_small = t.ppf((1 + confidence_level_small) / 2, df=len(small_class_scores)-1)

# Calculating confidence interval
confidence_interval_small = (
    sample_mean_small - t_score_small * standard_error_small,
    sample_mean_small + t_score_small * standard_error_small
)

print(f"Sample Mean: {sample_mean_small}")
print(f"Standard Error: {standard_error_small}")
print(f"T-Score: {t_score_small}")
print(f"Confidence Interval: {confidence_interval_small}")

# Plotting the data
plt.hist(small_class_scores, bins=5, color='lightgreen', alpha=0.7, label='Exam Scores (Small Class)')

# Plotting the confidence interval
plt.axvline(confidence_interval_small[0], color='blue', linestyle='dashed', linewidth=2, label='Confidence Interval')
plt.axvline(confidence_interval_small[1], color='blue', linestyle='dashed', linewidth=2)

plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('T-Distribution Confidence Interval')
plt.legend()
plt.show()
```

# 3. Proportion Confidence Interval

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Example data: survey responses (1 for satisfied, 0 for not satisfied)
customer_satisfaction = [1, 1, 1, 0, 1, 1, 1, 1, 0, 1]

# Calculating sample proportion
sample_proportion = np.mean(customer_satisfaction)

# Calculating standard error for proportions
standard_error_proportion = np.sqrt((sample_proportion * (1 - sample_proportion)) / len(customer_satisfaction))

# Setting confidence level
confidence_level_proportion = 0.95

# Calculating Z-score
z_score_proportion = norm.ppf((1 + confidence_level_proportion) / 2)

# Calculating confidence interval for proportions
confidence_interval_proportion = (
    sample_proportion - z_score_proportion * standard_error_proportion,
    sample_proportion + z_score_proportion * standard_error_proportion
)
print(f"Sample Proportion: {sample_proportion}")
print(f"Standard Error: {standard_error_proportion}")
print(f"Z-Score: {z_score_proportion}")
print(f"Confidence Interval: {confidence_interval_proportion}")

# Plotting the data
plt.bar(['Satisfied', 'Not Satisfied'], [sum(customer_satisfaction), len(customer_satisfaction) - sum(customer_satisfaction)], color=['green', 'red'], alpha=0.7, label='Customer Satisfaction')

# Plotting the confidence interval
plt.axhline(y=confidence_interval_proportion[0] * len(customer_satisfaction), color='blue', linestyle='dashed', linewidth=2, label='Confidence Interval')
plt.axhline(y=confidence_interval_proportion[1] * len(customer_satisfaction), color='blue', linestyle='dashed', linewidth=2)

plt.ylabel('Number of Customers')
plt.title('Proportion Confidence Interval')
plt.legend()
plt.show()
```

# 4. Bootstrap Confidence Interval

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: heights of individuals in a sample
heights_sample = [165, 170, 155, 175, 160]

# Number of bootstrap samples
num_bootstrap_samples = 1000

# Generate bootstrap samples
bootstrap_samples = [np.random.choice(heights_sample, size=len(heights_sample), replace=True) for _ in range(num_bootstrap_samples)]

# Calculate means for each bootstrap sample
bootstrap_means = [np.mean(bootstrap_sample) for bootstrap_sample in bootstrap_samples]

# Calculate confidence interval based on percentiles
confidence_interval_bootstrap = np.percentile(bootstrap_means, [2.5, 97.5])

print(f"Bootstrap Means: {bootstrap_means}")
print(f"Confidence Interval (Bootstrap): {confidence_interval_bootstrap}")

# Plotting the data
plt.hist(heights_sample, bins=10, color='lightcoral', alpha=0.7, label='Height Samples')

# Plotting the confidence interval
plt.axvline(confidence_interval_bootstrap[0], color='green', linestyle='dashed', linewidth=2, label='Confidence Interval')
plt.axvline(confidence_interval_bootstrap[1], color='green', linestyle='dashed', linewidth=2)

plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Bootstrap Confidence Interval')
plt.legend()
plt.show()
```

# 5. Pooled Confidence Interval

```python
import numpy as np
from scipy.stats import norm

# Example data: exam scores of two groups
group_1_scores = [75, 80, 85, 90, 92]
group_2_scores = [78, 85, 88, 92, 95]

# Calculating sample means
mean_group_1 = np.mean(group_1_scores)
mean_group_2 = np.mean(group_2_scores)

# Calculating standard deviations
std_group_1 = np.std(group_1_scores, ddof=1)
std_group_2 = np.std(group_2_scores, ddof=1)

# Calculating pooled standard deviation
pooled_std = np.sqrt(((len(group_1_scores) - 1) * std_group_1**2 + (len(group_2_scores) - 1) * std_group_2**2) / (len(group_1_scores) + len(group_2_scores) - 2))

# Calculating standard error for the pooled confidence interval
standard_error_pooled = pooled_std * np.sqrt(1/len(group_1_scores) + 1/len(group_2_scores))

# Setting confidence level
confidence_level_pooled = 0.95

# Calculating Z-score
z_score_pooled = norm.ppf((1 + confidence_level_pooled) / 2)

# Calculating confidence interval for the pooled means
confidence_interval_pooled = (
    mean_group_1 - mean_group_2 - z_score_pooled * standard_error_pooled,
    mean_group_1 - mean_group_2 + z_score_pooled * standard_error_pooled
)

print(f"Sample Means: Group 1 - {mean_group_1}, Group 2 - {mean_group_2}")
print(f"Pooled Standard Deviation: {pooled_std}")
print(f"Standard Error (Pooled): {standard_error_pooled}")
print(f"Z-Score: {z_score_pooled}")
print(f"Confidence Interval (Pooled): {confidence_interval_pooled}")
```
