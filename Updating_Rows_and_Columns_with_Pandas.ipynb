```python
import pandas as pd

Student = {"Id": ['cs-01', 'mth-02', 'eng-03', 'phy-04', 'mth-05', 'cs-06'],
           "Name": ["Fatima Khan", 'Noureen Ali', 'Salman Aslam', 'Sana Khan', 'Adam Jack', 'Ali Khan'],
           "subject": ['computer science', 'math', 'english', 'physics', 'statistics', 'programming'],
           "marks": [45, 90, 60, 37, 49, 50]}
data = pd.DataFrame(Student)
data
```

```python
data.rename(columns = {'Name': 'Full Name'})
```

```python
data.columns = data.columns.str.lower()
data
```

```python
data.loc[4]
```

```python
data.loc[4] = ['cs-05', 'Ali', 'Database', 47]
```

```python
data.loc[0, ['name', 'subject']]
```

```python
data.loc[0, ['name', 'subject']] = ['Farhan', 'Python']
```

```python
filt = data['marks'] < 50
filt
```

```python
data.loc[filt, 'marks'] = 'Fail'
data
```
