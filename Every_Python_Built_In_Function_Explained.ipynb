# input() — The “Tell Me Something… Anything” Function

```python
name = input("What’s your name? ")
```

```python
age = int(input("Age: "))
```

# print() — The Programmer’s Therapy Tool

```python
print("Debugging is just print statements plus emotional damage.")
```

```python
print("Hello", name, age)
```

```python
print(f"{name} is {age} years old.")
```

# abs() — Absolute Value Without the Drama

```python
abs(-5)  # → 5
```

# round() — The Peacemaker Between Fractions and Whole Numbers

```python
round(3.6)  # → 4
round(3.4)  # → 3
```

# divmod() — Two Answers for the Price of One

```python
divmod(10, 3)  # → (3, 1)
```

# pow() — Because ** Wasn’t Enough

```python
pow(2, 3)  # → 8
```

# Type Conversion Trio (int, float, complex)

```python
int(3.9)       # → 3
float("10")   # → 10.0
complex(2, 5)  # → (2+5j)
```

# str() — The Universal Converter

```python
str(100)  # → "100"
```

# format() — Your Text’s Makeover Artist

```python
"Hello, {}".format("Python")
```

# ascii() — Hide the Weird Characters

```python
ascii("café")  # → 'café'
```

# chr() & ord() — Character/Number Ping-Pong

```python
chr(65)   # → 'A'
ord('A')  # → 65
```

# len() — How Big Is This Thing?

```python
len([1,2,3])  # → 3
```

# sorted() — Finally Some Order

```python
sorted([3,1,2])  # → [1,2,3]
```

# reversed() — Flip It

```python
list(reversed([1,2,3]))  # → [3,2,1]
```

# enumerate() — Counting Like a Functional Adult

```python
for i, item in enumerate(['a','b','c'], start=1):
    print(i, item)
```

# zip() — The Matchmaker for Lists

```python
zip([1,2,3], ['a','b','c'])
```

# all() & any() — Group Truth Evaluators

```python
all([True, True])  # → True
any([False, True]) # → True
```

# max() & min() — Extremists

```python
max([1,2,3])  # → 3
min([1,2,3])  # → 1
```

# sum() — No Calculator Needed

```python
sum([1,2,3])  # → 6
```

# type() — What Are You, Exactly?

```python
type(3)  # → <class 'int'>
```

# id() — The Memory Name Tag

```python
isinstance(5, int)     # True
issubclass(bool, int)  # True
```

# help() — The Built‑In Documentation Genie

```python
help(str)
```
