## SMOTE Variants (When Standard SMOTE Isnâ€™t Enough)

- ADASYN (Adaptive Synthetic Sampling): Focuses on hard-to-learn minority samples. Generates more synthetic samples around minority samples that are near the decision boundary.
- SMOTE-ENN
Combines SMOTE (oversampling) + Edited Nearest Neighbours (removes noisy samples).
- SMOTE-TOMEK
Combines SMOTE + Tomek Links (removes overlapping borderline samples).
- SMOTE-NC
For datasets with both numerical AND categorical features.

# Technique 3: Under-sampling (Remove Majority Samples)
Keep all minority samples. Randomly remove the majority of samples to balance the dataset.

# Technique 4: Threshold Calibration (The Overlooked Gem)
Classification models output probabilities. You apply a threshold to convert probabilities to class predictions.
