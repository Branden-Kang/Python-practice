```python
# import pandas & numpy library
import pandas as pd
import numpy as np
```

```python
# Import pandas library
import pandas as pd
# Read the online file by the URL privded above, and assign it to variable "df"
other_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(other_path, header=None)
```

```python
# Show the first 5 rows using dataframe.head() method
print("The first 5 rows of the dataframe")
df.head(5)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel type","aspiration", "num-of-doors","body-style", "drive wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type", "num-of-cylinders", "engine-size","fuel system","bore","stroke","compression-ratio","horsepower", "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = headers
df.head(10)
```

# Data Wrangling


```python
df.replace('?', np.nan, inplace=True)
df.head(10)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df["stroke"].astype('float').mean(axis=0)
print("Average of Stroke", avg_stroke)
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_horsepower = df["horsepower"].astype("float").mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm = df["peak-rpm"].astype("float").mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df["num-of-doors"].value_counts()
```

```python
df["num-of-doors"].value_counts().idmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)
# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]
# check your transformed data
df.head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]
# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)
# check your transformed data
df.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()
# show the scaled columns
df[["length","width","height"]].head()
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
dummy_variable_1 = pd.get_dummies(df["fuel type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1"
df = pd.concat([df, dummy_variable_1], axis=1)
# drop original column "fuel-type" from "df"
df.drop("fuel type", axis = 1, inplace=True)
df.head()
```

```python
# get indicator variables of aspiration and assign it to data frame "dummy_variale_2"
dummy_variable_2 = pd.get_dummies(df["aspiration"])
# change column names for clarity
dummy_variable_2.rename(columns={"std":"aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)
# show first 5 instances of data frame "dummy_variable_1"
dummy_variable_2.head()
```

```python
# merge the new dataframe to the original dataframe
df = pd.concat([df, dummy_variable_2], axis=1)
# drop original column "aspiration" from "df"
df.drop("aspiration", axis=1, inplace=True)
# Display new data frame
df.head()
```
