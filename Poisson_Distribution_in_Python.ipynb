```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from numpy.random import randn
from numpy import random as rn
from scipy.stats import poisson
import scipy.stats as si
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import Image
%matplotlib inline

Lambda = 12
```

```python
## Monte Carlo Model
Simulations = 1000
f = np.arange(1,Simulations+1)
df = pd.DataFrame(data=f)
df['Poisson'] = np.random.poisson(Lambda, Simulations)
df.drop([0], axis=1, inplace=True)
df
```

```python
V = []
Simulations = 10
Trails = 1000
for j in range(0,Simulations):
    f = np.arange(1,Trails+1)
    df = pd.DataFrame(data=f)
    df['Poisson'] = np.random.poisson(Lambda,Trails)
    df.drop([0], axis=1, inplace=True)
    countif = len(df[(df['Poisson'] == 15)])
    count = len(df['Poisson'])
    qq = countif/count
    V.append(qq)
```
