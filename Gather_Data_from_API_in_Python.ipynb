```python
import pandas as pd
import requests
```

```python
URL = 'https://api.nobelprize.org/2.1/laureates'
params = {'limit':'1000', 'offset':'1'}
response = requests.get(URL,params=params)
```

```python
print(response)
```

```python
data = response.json()['laureates']
print(data)
```

```python
from pandas.io.json import json_normalize
df = json_normalize(data)
df
```

```python
import pandas as pd
from pandas.io.json import json_normalize
import requests
URL = 'https://api.nobelprize.org/2.1/laureates'
params = {'limit':'1000', 'offset':'1'}
response = requests.get(URL,params=params)
data=response.json()['laureates']
df = json_normalize(data)
print(df)
```
