```python
import pandas as pd
```

```python
df = pd.read_csv('survey_results_public.csv', index_col='ResponseId')
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.sort_values(by ='Country')
```

```python
df.sort_values(by = 'Country', ascending=False)
```

```python
df.sort_values(by = ['Country', 'ConvertedCompYearly'])
```

```python
df.sort_values(by = ['Country', 'ConvertedCompYearly'], ascending = [True, False])
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```

```python
df.sort_index(axis='columns')
```
