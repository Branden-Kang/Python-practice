```python
import pandas as pd
df = pd.DataFrame({'A': ['a1', 'a2', 'a3'], 

                   'B': ['b1', 'b2', 'b3'], 

                   'C': ['c1', 'c2', 'c3']}) 
df
```

```python
df_stacked = df.stack()
df_stacked
```

```python
df_unstacked = df_stacked.unstack()
df_unstacked
```
