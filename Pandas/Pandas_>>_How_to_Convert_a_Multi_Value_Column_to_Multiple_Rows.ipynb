```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Kevin", "Jack", "Mary"],
    "sex": ["M", "M", "F"],
    "hobbies": ["Writing,Chess", "Drawing,Coding", "Music,Reading"]
}, index=["K", "J", "M"])

df
```

```python
df["hobbies"] = df["hobbies"].str.split(",") 
df
```

```python
df = df.explode("hobbies") 
df
```
