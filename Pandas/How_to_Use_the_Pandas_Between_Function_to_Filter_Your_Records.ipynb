```python
import pandas as pd
fruit_data = {
    "fruit": ["apples", "oranges", "bananas", "strawberries"],
    "count": [5, 7, 3, 2],
    "price": [3.5, 4.5, 2.1, 3.0],
    "date_of_sale": ["2020-08-04", "2020-08-07", "2020-08-08", "2020-06-08"],
}
df = pd.DataFrame(fruit_data)
display(df)
```

```python
df[df['count'].between(4, 7, inclusive='both')]
```

```python
df[df['count'].between(2, 7, inclusive='neither')]
```

```python
df[df['date_of_sale'].between('2020-06-08','2020-08-07', inclusive='both')]
```
