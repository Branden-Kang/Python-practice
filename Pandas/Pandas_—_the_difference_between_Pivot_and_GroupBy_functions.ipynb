```python
import pandas as pd
df = pd.DataFrame({'Country': ['USA', 'USA', 'China', 'China'], 

                   'Year': ['2000', '2001', '2000', '2001'], 

                   'Population': [281, 282, 1260, 1261]}) 
df
```

```python
df.pivot(index='Country', columns='Year', values='Population')
```

```python
grp = df.groupby(by=['Country','Year'])
grp
```

```python
df.groupby(by=['Country','Year']).sum()
```
