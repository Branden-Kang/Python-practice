```python
import pandas as pd
```

```python
df = pd.DataFrame({"Col1": [10, 20, 15, 30, 45],"Col2": [13, 23, 18, 33, 48],"Col3": [17, 27, 22, 37, 52]},index=pd.date_range("2022-01-01", "2022-01-05"))
df
```

```python
df.shift(periods=1)
```

```python
df.shift(periods=2, freq='D')
```

```python
df.shift(periods=3, fill_value=0)
```
