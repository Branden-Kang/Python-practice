# 2. YellowBrick

```python
# YellowBrick installation
! pip install yellowbrick

# Download the room occupancy dataset
from yellowbrick.datasets import load_occupancy
X, y = load_occupancy()
```

```python
from yellowbrick.features import JointPlotVisualizer

visualizer = JointPlotVisualizer(columns=['temperature', 'light' ])
visualizer.fit_transform(X, y)
visualizer.show()
```

```python
from yellowbrick.features import parallel_coordinates
visualizer = parallel_coordinates(X, y, normalize="standard")
```

```python
from yellowbrick.features import pca_decomposition
visualizer = pca_decomposition(X, y)
```

# 4. SHAP

```python
! pip install shap

# xgb_classifier is a scikit-learn classifier that has already been trained
import shap
explainer = shap.TreeExplainer(xgb_classifier)
shap_values = explainer.shap_values(X)

# X is a pandas Dataframe with the input features 
# y is a pandas Series with the label per submission
# Feature Importance
shap.summary_plot(shap_values, X.values, plot_type="bar", class_names= list(y.reset_index()['Label'].drop_duplicates()), feature_names = X.columns)

# Beeswarm plot
shap.summary_plot(shap_values, X.values, feature_names = X.columns)
```
