# 1. clip

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    np.random.randint(-10, 10, size=(5, 5)), 
    columns=list("ABCDE")
)

df
```

```python
df.clip(lower=-4)
```

```python
df.clip(lower=-4, upper=4)
```

# 2. diff

```python
df
```

```python
# difference between consecutive rows
df.diff()
```

```python
# difference between the row before
df.diff(periods=2)
```

```python
# difference between consecutive columns
df.diff(axis=1)
```

# 3. get_dummies

```python
import pandas as pd

df = pd.DataFrame(

    {
        "age": [45, 53, 60, 42, 34],
        "gender": ["Female", "Male", "Female", "Female", "Male"]
    }
)

df
```

```python
pd.get_dummies(df)
```

# pd.get_dummies(df)

```python
df_new = pd.get_dummies(df)
df_new
```

```python
pd.from_dummies(df_new[["gender_Female", "gender_Male"]])
```

# 5. transform

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
   np.random.randint(0, 10, size=(5, 3)), 
   columns=list("ABC")
)

df
```

```python
df.transform([np.sqrt, np.square])
```
