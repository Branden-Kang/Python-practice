# map, filter, and reduce with lambdas

````
foo = lambda n: n**2 if n%2 == 0 else n**3
````

````
foo(5)
````

````
def addition(x):
  return x+100

def transform(x):
  if x%2 == 0:
    return x*2
  else:
    return x + 5

functions = [addition, transform]
````

````
functions
````

````
my_var = 50
````

````
for function in functions:
  print(function(my_var))
````

````
def maximize(var):
  return var*100

col = [10, 15, 19]

results = map(maximize, col)
````

````
print(results)
````

map() function is a lazy evaluation

````
results = list(map(maximize, col))
print(results)
````

````
for result in results:
  print(result)
````

````
col = [10, 15, 19]
results = map(lambda x: x*10, col)
print(list(results))
````

filter()

````
col = [20, 128, 880, 9, 17, 8, 10, 55]
````

````
filtered = filter(lambda x: x%2 ==0 and x > 25, col)
````

````
for filter in filtered:
  print(filter)
````

reduce() is like <b>accumulator</b>.

````
from functools import reduce

example = [1, 10, 8, 55, 200, 35]

result = reduce(lambda x, y: x*y+10, example)
result
````

# Context managers

````
# with open("sample.txt","w") as file:
#   # logic
````

````
import datetime

class DataManage():
  def __init__(self):
    self.file = None
  
  def __enter__(self):
    now = str(datetime.now()).split(".")[0].replace(" ", "-").replace(":", "-")
    filename = now + "-DATA.txt"
    self.file = open(filename, "w")
    return self.file
  
  def __exit__(self, exc_type, exc_value, exc_traceback):
    self.file.close()
    print("I exited")

with DataManger() as data:
  data.write("hello")
````

````
class validate_positive():
  def __init__(self, func):
    self.function = func

  def __call__(self, *args, **kwargs):
    for arg in args:
      if arg < 0:
        raise Exception("Invalid Number!")
    return self.function(*args, **kwargs)

@validate_positive
def complicated_computation(x, y, z):
  return x**2 + y*10 - z

print(complicated_computation(-1, 15, 7))
````

# Generators

````
def squares(a, b):
    i = a
    while i < b:
        yield i**2
        i += 1
````

````
for num in squares(5, 10):
    print(num)
````

````
sequence = squares(5, 10)
````

````
print(next(sequence))
print(next(sequence))
````
