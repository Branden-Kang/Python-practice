# 1. @staticmethod

```python
class Calculator:
    @staticmethod
    def add(a, b):
        return a + b

# Called directly without instantiation
result = Calculator.add(5, 7)
print(result) # Output: 12
```

# 2. @classmethod

```python
class Person:
    # A class-level attribute (class variable)
    species = "Homo sapiens"

    def __init__(self, name, age):
        self.name = name  # Instance attribute
        self.age = age    # Instance attribute

    @classmethod
    def get_species(cls):
        """A class method to access a class variable."""
        return cls.species

    @classmethod
    def from_birth_year(cls, name, birth_year):
        """An alternative constructor (factory method)."""
        from datetime import date
        age = date.today().year - birth_year
        return cls(name, age) # This calls the __init__ of the specific class (or subclass)

# Accessing the class method
print(Person.get_species()) # Output: Homo sapiens

# Using the alternative constructor
person1 = Person.from_birth_year("Alice", 1990)
print(f"{person1.name} is {person1.age} years old.")
```

# 3. @property

```python
class Person:
    def __init__(self, name):
        self._name = name

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not value:
            raise ValueError("Name cannot be empty")
        self._name = value

# Usage
p = Person("Alice")
print(p.name)  # Accesses method like an attribute
p.name = "Bob" # Uses the setter
```

# 4. @functools.lru_cache

```python
from functools import lru_cache

@lru_cache(maxsize=128)
def fib(n):
    if n < 2:
        return n
    return fib(n-1) + fib(n-2)
```

# 5. @dataclass

```python
from dataclasses import dataclass

@dataclass
class User:
    id: int
    name: str
    active: bool = True

user1 = User(1, "Alice")
print(user1)  # Output: User(id=1, name='Alice', active=True)
```

# 6. @contextmanager

```python
from contextlib import contextmanager

@contextmanager
def managed_resource():
    print("Step 1: Setup (e.g., open file)")
    try:
        yield "Resource Data"  # This is what 'as res' gets
    finally:
        print("Step 3: Cleanup (e.g., close file)")

# Usage
with managed_resource() as res:
    print(f"Step 2: Using {res}")
```

# 7. @wraps

```python
import functools

def my_decorator(func):
    """
    A custom decorator.
    """
    @functools.wraps(func)  # Apply @wraps to the wrapper function, passing the original 'func'
    def wrapper(*args, **kwargs):
        """
        The wrapper function for the decorator.
        """
        print(f"Something is happening before {func.__name__!r} is called.")
        result = func(*args, **kwargs)
        print(f"Something is happening after {func.__name__!r} is called.")
        return result
    return wrapper

@my_decorator
def say_hello(name):
    """
    A simple function that says hello.
    """
    print(f"Hello, {name}!")

# Without @wraps, help(say_hello) would show wrapper's docstring and name
help(say_hello)
# Output now correctly shows the original function's info:
# Help on function say_hello in module __main__:
# say_hello(name)
#     A simple function that says hello.
```

# 8. @retry (custom decorator)

```python
import time

def retry(times):

    def decorator(func):

        def wrapper(*args, **kwargs):

            for i in range(times):
                try:
                    return func(*args, **kwargs)
                except Exception:
                    time.sleep(1)

        return wrapper

    return decorator


@retry(3)
def api_call():
    print("Calling API")
```

# 9. @timer (performance decorator)

```python
import time
from functools import wraps

def timer(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()

        print("Time:", end - start)

        return result

    return wrapper
```

# 10. @login_required (real-world style)

```python
def login_required(func):

    def wrapper(user):

        if not user.get("logged_in"):
            raise Exception("Login required")

        return func(user)

    return wrapper


@login_required
def dashboard(user):
    print("Welcome")
```
