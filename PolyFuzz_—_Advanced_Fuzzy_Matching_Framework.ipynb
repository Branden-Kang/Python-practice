```python
from polyfuzz import PolyFuzz
import config

model = PolyFuzz("EditDistance")
model.match(config.list1, config.list2)

print(model.get_matches())
```

```python
from polyfuzz import PolyFuzz
import config

model = PolyFuzz("EditDistance")
model.match(config.list1, config.list2)
model.group("EditDistance")
print(model.get_matches())
```

```python
from polyfuzz import PolyFuzz
import config

model = PolyFuzz("EditDistance")
model.match(config.list1, config.list2)
model.group("EditDistance")
print(model.get_clusters())
```
