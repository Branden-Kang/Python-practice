```python
import pandas as pd
```

```python
employees = ["Bob", "Mark", "Jane", "Patrick"]

data = pd.DataFrame(employees)
data
```

```python
employees = ["Bob", "Mark", "Jane", "Patrick"]

data = pd.DataFrame(employees, columns=["First Name"])
data
```

```python
e_first_names = ["Bob", "Mark", "Jane", "Patrick"]
e_last_names = ["Doe", "Markson", "Swift", "Johnson"]

data = pd.DataFrame(
 data=list(zip(e_first_names, e_last_names)),
 columns=["First Name", "Last Name"]
)
data
```

```python
e_first_names = ["Bob", "Mark", "Jane", "Patrick"]
e_last_names = ["Doe", "Markson", "Swift", "Johnson"]
e_emails = ["bdoe@company.com", "mmarkson@company.com", "jswift@company.com", "pjohnson@company.com"]

data = pd.DataFrame(
 data=list(zip(e_first_names, e_last_names, e_emails)),
 columns=["First Name", "Last Name", "Email"]
)
data
```

```python
employees = [
 ["Bob", "Doe", "bdoe@company.com"],
 ["Mark", "Markson", "mmarkson@company.com"],
 ["Jane", "Swift", "jswift@company.com"],
 ["Patrick", "Johnson", "pjohnson@company.com"]
]
 
data = pd.DataFrame(
 data=employees,
 columns=["First Name", "Last Name", "Email"]
)
data
```

```python
employees = {
 "First Name": ["Bob", "Mark", "Jane", "Patrick"],
 "Last Name": ["Doe", "Markson", "Swift", "Johnson"],
 "Email": ["bdoe@company.com", "mmarkson@company.com", "jswift@company.com", "pjohnson@company.com"]
}

data = pd.DataFrame(employees)
data
```

```python
employees = [
 ["Bob", "Doe", "bdoe@company.com"],
 ["Mark", "Markson", "mmarkson@company.com"],
 ["Jane", "Swift", "jswift@company.com"],
 ["Patrick", "Johnson", "pjohnson@company.com"]
]

data = pd.DataFrame.from_records(employees)
data
```

```python
employees = [
 ["Bob", "Doe", "bdoe@company.com"],
 ["Mark", "Markson", "mmarkson@company.com"],
 ["Jane", "Swift", "jswift@company.com"],
 ["Patrick", "Johnson", "pjohnson@company.com"]
]

data = pd.DataFrame.from_records(employees, columns=["First Name", "Last Name", "Email"])
data
```
